```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Newton-Raphson Method vs np.sqrt(x) 

```python
def deltax(f, x):
    return abs(0. - f(x))

def NewtonRaphson_Method(f, df, x0, toll):
    delta = deltax(f, x0)
    count = 0
    while delta > toll:
        x0 = x0 - f(x0)/df(x0)
        delta = deltax(f, x0)
        count += 1
    error = abs(2.0 - x0)
    return x0, error, count 
    
```

```python
def f(x):
    ret = lambda y: y**2 - x
    return ret

def df(y):
    return 2 * y

print NewtonRaphson_Method(f(4), df, 0.5, 1.0)
print NewtonRaphson_Method(f(4), df, 0.5, 1e-1)
print NewtonRaphson_Method(f(4), df, 0.5, 1e-2)
print NewtonRaphson_Method(f(4), df, 0.5, 1e-3)
print NewtonRaphson_Method(f(4), df, 0.5, 1e-15)
```

```python
sol = []
x0 = .5
for x in np.arange(0.2,20.2,0.1):
    my_f = f(x)
    my_df = df
    my_sol = NewtonRaphson_Method(my_f, my_df, x0, 1e-5)
    sol.append(my_sol[0])
```

```python
x = np.linspace(0.2,20.2,100)

plt.figure(figsize=(10,7))
plt.title('Newton-Raphson Method vs np.sqrt(x)')
plt.plot(x, np.sqrt(x), linewidth=4, label='exact')
plt.plot(np.arange(0.2,20.2,0.1), sol, "o", markersize = 2, label='New-Rap')
plt.legend(loc='upper left')
plt.savefig('nr.png')
plt.show()
```
